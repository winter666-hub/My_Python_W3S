### Linestyle
### 선 스타일 - 점으로 표현

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

### Use a dashed line
### 점선으로 표현

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dashed')
plt.show()
```

### linestyle을 ls로 쓸 수 있다. dotted를 : 로 쓸 수 있다. dashed를 -- 로 쓸 수 있다

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls = ':')
plt.show()
```

### Line Color
### 선 색상을 빨간색으로 설정

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'r')
plt.show()
```

### Plot with a beautiful green line
### 선 색상을 초록색 및 핫핑크로 설정

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, c = '#4CAF50') # color도 c로 표현 가능
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, c = 'hotpink')
plt.show()
```

### Line Width
### 선의 너비를 설정

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

### Multiple Lines
### 다중의 선을 생성

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

### Draw two lines by specifiyng the x- and y-point values for both lines
### 두 점 x, y값을 지정하여 두 선을 그리기

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```
