### Creating Pie Charts
### 파이 차트 만들기 pie()

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15]) # 합계가 100

plt.pie(y)
plt.show()
# 시작은 원의 중심기준 우측 즉, x축부터 시계 반대방향으로
```

### Labels
### labels 매개변수를 사용하여 레이블 추가

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

### Start Angle
### startangle 매개변수로 시작 각도 지정 

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

### Explode
### explode 매개벼수로 하나의 파이 조각을 떼어내기

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

### Shadow
### shadows 매개변수로 그림자 추가

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

### Colors
### colors 매개변수로 각 파이의 색상 설정

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "blue", "#4CAF50"] # blue대신 b로 가능
# r, g, b 등등
plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

### Legend
### legned() 함수로 각 파이에 설명 목록 추가

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

### Legend With Header
### 설명 목록에 헤더 추가

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```
