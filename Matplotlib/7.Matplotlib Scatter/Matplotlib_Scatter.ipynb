### Creating Scatter Plots
### 산점도 만들기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12 , 9, 6]) # 자동차의 연식
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]) # 자동차의 속도

plt.scatter(x, y) # 그래프의 산점도 반환
plt.show()
```

### Compare Plots
### 산점도 비교

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)


x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

### Colors
### 산점도의 색상 변경

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')


x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = 'green')

plt.show()
```

### Color Each Dot
### 각 점을 색칠

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

### ColorMap
### 컬러맵을 사용하는 방법 cmap, viridis

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100]) # 컬러맵 색 지정
# viridis기준. 낮은 값은 보라색 계열, 중간 값은 초록 계열, 높은 값은 노란색 계열
plt.scatter(x, y, c = colors, cmap='viridis') # 각 점의 색을 숫자로 지정.
# cmap은 숫자를 어떤 색상에 매핑할 지 결정

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100]) # 컬러맵 색 지정
# plasma기준. 보라 -> 빨강 -> 노랑
plt.scatter(x, y, c = colors, cmap='plasma') # 각 점의 색을 숫자로 지정.
# cmap은 숫자를 어떤 색상에 매핑할 지 결정

plt.show()
```

### 컬러맵 표현

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

### Size
### 인수를 사용하여 점의 크기 변경

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

### Alpha
### alpha를 사용해 마커의 투명도를 조정

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha = 0.5)

plt.show()
```

### Combine Color Size and Alpha
### 색상의 범위와 알파를 결합

```python
x = np.random.randint(100, size =(100))
y = np.random.randint(100, size = (100))

colors = np.random.randint(100, size = (100))
sizes = 10 * np.random.randint(100, size = (100))

plt.scatter(x, y, c = colors, s = sizes, alpha = 0.5, cmap = 'nipy_spectral')

plt.colorbar()

plt.show()
```
