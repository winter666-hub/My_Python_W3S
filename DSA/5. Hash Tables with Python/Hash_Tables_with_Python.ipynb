### Hash Table
### 해시 테이블
### 해시 테이블은 빠르게 작업할 수 있도록 설계된 데이터 구조.
### 해시 테이블이 배열이나 연결 리스트보다 선호되는 이유는 데이터 검색, 삽입, 삭제를 대량의 데이터라도 빠르게 수행할 수 있기 때문이다.

### 해시 테이블 생성 
### 1. 빈 리스트를 만는다.
### 2. 해시 함수를 생성.
### 3. 해시 함수를 사용하여 요소를 삽입
### 4. 해시 함수를 사용하여 요소를 검색
### 5. 충돌 처리


### Step 1: Create an Empty List
### 1. 빈 리스트 생성

```python
# 빈 요소 10개가 있는 리스트 생성
my_list = [None, None, None, None, None, None, None, None, None, None]
```

### Step 2: Create a Hash Function
### 2. 해시 함수 생성

```python
def hash_function(value):
    sum_of_char = 0
    for char in value: # Bob을 한 글자씩 반복
        # 유니코드로 변환 후 연산 66 + 111 + 98 = 275
        sum_of_char += ord(char)
    # 275 % 10 = 5    
    return sum_of_char % 10

print("'Bob'의 해시 코드:", hash_function('Bob'))
```

### Step 3: Inserting an Element
### 3. 요소 삽입

```python
my_list = [None, None, None, None, None, None, None, None, None, None]

def hash_function(value):
    sum_of_char = 0
    for char in value: 
        sum_of_char += ord(char)
    return sum_of_char % 10

def add(name):
    # name을 해싱하여 인덱스를 얻는다.
    index = hash_function(name)
    # 해시 테이블의 해당 인덱스 위치에 name을 저장
    my_list[index] = name
    
add('Bob')
print(my_list)
```

### 이름 추가

```python
my_list = [None, None, None, None, None, None, None, None, None, None]

def hash_function(value):
    sum_of_char = 0
    for char in value: 
        sum_of_char += ord(char)
    return sum_of_char % 10

def add(name):
    # name을 해싱하여 인덱스를 얻는다.
    index = hash_function(name)
    # 해시 테이블의 해당 인덱스 위치에 name을 저장
    my_list[index] = name
    
add('Bob')
add('Pete')
add('Jones')
add('Lisa')
add('Siri')
print(my_list)
```

### Step 4: Looking up a name
### 4. 이름 찾기

```python
my_list = [None, None, None, None, None, None, None, None, None, None]

def hash_function(value):
    sum_of_char = 0
    for char in value: 
        sum_of_char += ord(char)
    return sum_of_char % 10

def add(name):
    # name을 해싱하여 인덱스를 얻는다.
    index = hash_function(name)
    # 해시 테이블의 해당 인덱스 위치에 name을 저장
    my_list[index] = name

def contains(name):
    index = hash_function(name)
    return my_list[index] == name
   
add('Bob')
add('Pete')
add('Jones')
add('Lisa')
add('Siri')
print(my_list)

print("'Pete'가 해시 테이블에 있는가?:", contains('Pete'))
```
