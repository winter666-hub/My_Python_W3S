### Step 2: Create a Hash Function
### 2. 해시 함수 생성

```python
def hash_function(value):
    sum_of_char = 0
    for char in value:
        sum_of_char += ord(char)
        
    return sum_of_char % 10

print("'Bob'의 해시 코드:", hash_function('Bob'))
```
