### Create Algorithms
### 알고리즘 생성

```python
my_array = [7, 12, 9, 4, 11, 8, 4]
minVal = my_array[0]

for i in my_array:
    if i < minVal:
        minVal = i

print("Lowest value :", minVal)
```

### 시간 복잡도는 알고리즘 입력값의 크기에 따라 연산 횟수가 얼마나 늘어나는지을 말한다, 위의 코드는 배열이 크면 클 수록 더욱 시간이 오래 걸리고 배열에 1000개의 값이 있다면 루프는 1000번이 실행되어야 한다.
